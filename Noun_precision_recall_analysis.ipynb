```python
def noun_count(comp):
    extra_nouns = 0
    missing_nouns = 0
    correct_nouns = 0
    total_nouns = 0
    for i in comp:
        #print(i)
        hyp_words = nltk.word_tokenize(i[0])
        ref_words = nltk.word_tokenize(i[1])
        hyp_tagged = nltk.pos_tag(hyp_words) 
        ref_tagged = nltk.pos_tag(ref_words) 
        hyp_nouns = [w[0] for w in hyp_tagged if w[1] == "NN"]
        ref_nouns = [w[0] for w in ref_tagged if w[1] == "NN"]
        #print(hyp_nouns)
        #print(ref_nouns)
        for n in hyp_nouns:
            if n not in ref_nouns:
                extra_nouns +=1
        for n in ref_nouns:
            if n not in hyp_nouns:
                missing_nouns +=1
            else:
                correct_nouns +=1
            total_nouns += 1
    print("extra:" + str(extra_nouns))
    print("missing:" + str(missing_nouns))
    print("correct:" + str(correct_nouns))
    print("total:" + str(total_nouns))
```

```python
noun_count(attn_comparisons)
```
